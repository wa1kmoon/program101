```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from matplotlib.gridspec import GridSpec
#from matplotlib.pyplot import MultipleLocator
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
from matplotlib.colors import BoundaryNorm
from matplotlib.ticker import MaxNLocator
import astropy.io.fits as fits
from astropy.wcs import WCS
from astropy import units as u
```

## 1d spec

```python
flux1d=[]
wl1d=[]
with open('normalized.txt','r') as ns:
    for line in ns.readlines():
        w,f=line.strip(' \n').split()
        flux1d.append(float(f))
        wl1d.append(float(w))

flux1d=np.array(flux1d)
wl1d=np.array(wl1d)
```

```python
n=int(len(wl1d)/6)+1

wls = [wl1d[i*(n+1):(i+1)*(n+1)] for i in range(5)]
wls.append(wl1d[5*(n+1):])

fls = [flux1d[i*(n+1):(i+1)*(n+1)] for i in range(5)]
fls.append(flux1d[5*(n+1):])
```

## 2d spec

```python
hdu_uvb=fits.open('FLUX_MERGE2D_UVB_UVB_1x2_100k.fits')[0]
hdu_vis=fits.open('FLUX_MERGE2D_VIS_VIS_1x2_100k.fits')[0]
```

```python
wcs_uvb=WCS(hdu_uvb.header)
data_uvb=hdu_uvb.data[35:76,:]
shy_uvb,shx_uvb=data_uvb.shape
wll_uvb, wlu_uvb=(wcs_uvb.pixel_to_world(0,int(shy_uvb/2))[0].value*10,wcs_uvb.pixel_to_world(shx_uvb-1,int(shy_uvb/2))[0].value*10) # 波长上下限
q2_uvb = wcs_uvb.pixel_to_world(int(shx_uvb/2),int(shy_uvb/2))[1].value

wcs_vis=WCS(hdu_vis.header)
data_vis=hdu_vis.data[35:76,:]
shy_vis,shx_vis=data_vis.shape
wll_vis, wlu_vis=(wcs_vis.pixel_to_world(0,int(shy_vis/2))[0].value*10,wcs_vis.pixel_to_world(shx_vis-1,int(shy_vis/2))[0].value*10) # 波长上下限
q2_vis = wcs_vis.pixel_to_world(int(shx_vis/2),int(shy_vis/2))[1].value
```

```python
def wcs2pix_uvb(wavel,q2=q2_uvb):
    pix = int(wcs_uvb.world_to_pixel(wavel * u.Angstrom, q2 * u.arcsec)[0])
    return pix

def wcs2pix_vis(wavel,q2=q2_vis):
    pix = int(wcs_vis.world_to_pixel(wavel * u.Angstrom, q2 * u.arcsec)[0])
    return pix


def pix2wcs_uvb(cx, cy=shy_uvb/2):
    wavel = wcs_uvb.pixel_to_world(cx,cy)[0].value*10
    return wavel

def pix2wcs_vis(cx, cy=shy_vis/2):
    wavel = wcs_vis.pixel_to_world(cx,cy)[0].value*10
    return wavel
```

```python
# 创建2d光谱的横坐标:从像素横坐标转换到波长
wl_uvb=np.array(list(map(pix2wcs_uvb, np.arange(0,shx_uvb,1))))
wl_vis=np.array(list(map(pix2wcs_vis, np.arange(0,shx_vis,1))))
```

```python
# 2d光谱纵坐标
cy_uvb=np.arange(0,shy_uvb,1)  # +1 for pcolormesh not warning
cy_vis=np.arange(0,shy_vis,1)  # +1 for pcolormesh not warning
```

```python
# UVB取全部, VIS取部分
sp2ds=[]
wl2ds=[]
for wl in wls:
    st = wl[0] # 当前波长段的最短波长
    en = wl[-1] # 当前波长段的最长波长
    #print(st,en)
    if en < wlu_uvb:
        st_px = int(wcs_uvb.world_to_pixel(st * u.Angstrom, q2_uvb * u.arcsec)[0])
        en_px = int(wcs_uvb.world_to_pixel(en * u.Angstrom, q2_uvb * u.arcsec)[0])
        sp2ds.append(data_uvb[:,st_px:en_px])
        wl2ds.append(wl_uvb[st_px:en_px])
        #print(st_px,en_px)
    elif st < wlu_uvb < en: # 如果波长段内包括uvb的波长上限,则合并uvb的部分和vis的部分
        st_px = int(wcs_uvb.world_to_pixel(st * u.Angstrom, q2_uvb * u.arcsec)[0])
        div_px_uvb = int(wcs_uvb.world_to_pixel(wlu_uvb * u.Angstrom, q2_uvb * u.arcsec)[0])
        div_px_vis = int(wcs_vis.world_to_pixel(wlu_uvb * u.Angstrom, q2_vis * u.arcsec)[0])
        en_px = int(wcs_vis.world_to_pixel(en * u.Angstrom, q2_vis * u.arcsec)[0])
        div_uvb = data_uvb[:,st_px:div_px_uvb]
        div_vis = data_vis[:,div_px_vis:en_px]
        sp2ds.append(np.hstack((div_uvb, div_vis)))
        wl2ds.append(np.hstack((wl_uvb[st_px:div_px_uvb], wl_vis[div_px_vis:en_px])))
        #print(st_px, div_px_uvb, div_px_vis, en_px)
    else:
        st_px = int(wcs_vis.world_to_pixel(st * u.Angstrom, q2_vis * u.arcsec)[0])
        en_px = int(wcs_vis.world_to_pixel(en * u.Angstrom, q2_vis * u.arcsec)[0])
        sp2ds.append(data_vis[:,st_px:en_px])
        wl2ds.append(wl_vis[st_px:en_px])
        #print(st_px,en_px)
```

```python
# UVB取部分, VIS取全部
sp2ds1=[]
wl2ds1=[]
for idx, wl in enumerate(wls):
    st = wl[0] # 当前波长段的最短波长
    en = wl[-1] # 当前波长段的最长波长 
    #print(st,en)
    if en < wlu_uvb:
        st_px = int(wcs_uvb.world_to_pixel(st * u.Angstrom, q2_uvb * u.arcsec)[0])
        en_px = int(wcs_uvb.world_to_pixel(en * u.Angstrom, q2_uvb * u.arcsec)[0])
        sp2ds1.append(data_uvb[:,st_px:en_px])
        wl2ds1.append(wl_uvb[st_px:en_px]) # +1 for pcolormesh not warning
        #print(st_px,en_px)
    else:
        st_px = int(wcs_vis.world_to_pixel(st * u.Angstrom, q2_vis * u.arcsec)[0])
        en_px = int(wcs_vis.world_to_pixel(en * u.Angstrom, q2_vis * u.arcsec)[0])
        sp2ds1.append(data_vis[:,st_px:en_px])
        if idx == len(wls)-1:
            wl2ds1.append(wl_vis[st_px:en_px]) # -1 for pcolormesh not warning
        else:
            wl2ds1.append(wl_vis[st_px:en_px]) # +1 for pcolormesh not warning
#        print(st_px,en_px)

#figt=plt.figure(figsize=(30,40))        
#plt.imshow(sp2ds1[2],cmap='Spectral_r', vmin=-2.e-17, vmax=2.e-18)
```

#### Making levels using Norms

```python
# make these smaller to increase the resolution
dx, dy = 0.05, 0.05
```

```python
slice(1,5+dx,dx) # 切片对象, 用于对数组进行切片截取, 也了用于创建网格
# generate 2 2d grids for the x & y bounds
y, x = np.mgrid[slice(1, 5 + dy, dy),
                slice(1, 5 + dx, dx)]
```

```python
z = np.sin(x)**10 + np.cos(10 + y*x) * np.cos(x)
```

```python
from matplotlib.colors import BoundaryNorm
from matplotlib.ticker import MaxNLocator
import matplotlib.colors as colors
z = z[:-1, :-1]
levels = MaxNLocator(nbins=15).tick_values(z.min(), z.max())
```

```python
cmap = plt.get_cmap('PiYG')
norm = BoundaryNorm(levels, ncolors=cmap.N, clip=True)


fig, (ax0, ax1) = plt.subplots(nrows=2)

im = ax0.pcolormesh(x, y, z, cmap=cmap, norm=norm)
fig.colorbar(im, ax=ax0)
ax0.set_title('pcolormesh with levels')
```
